```python
import math
```

```python
def run_calculator():
    while True: # [조건] 계속 계산할 것인지 묻기 위한 전체 루프

        # 1. 첫 번째 정수 입력할 수 있도록 만들기
        while True:
            try:
                num1 = int(input("첫 번째 정수를 입력하세요: "))
                break # 정수 입력 성공 시 루프 탈출
            except ValueError: # [조건] ValueError 처리
                print("잘못된 입력입니다. 정수를 입력해주세요.")

        # 2. 연산자 입력할 수 있도록 만들기
        operator = input("연산자를 입력하세요 (+, -, *, /, **): ")

        # 3. 두 번째 정수 입력할 수 있도록 만들기 (정수일 때까지 반복)
        while True:
            try:
                num2 = int(input("두 번째 정수를 입력하세요: "))
                break
            except ValueError:
                print("잘못된 입력입니다. 정수를 입력해주세요.")

        # 4. 연산 수행 및 결과 출력
        try:
            if operator == '+':
                result = num1 + num2
            elif operator == '-':
                result = num1 - num2
            elif operator == '*':
                result = num1 * num2
            elif operator == '/':
                # [조건] ZeroDivisionError 처리 필요 (0으로 나눴을 경우에)
                result = num1 / num2
            elif operator == '**':
                # [조건] math 모듈을 사용하여 제곱 연산 수행
                result = math.pow(num1, num2)
            else:
                # [조건] 지원하지 않는 연산자 처리
                print(f"에러: '{operator}'는 지원하지 않는 연산자입니다.")
                continue # 다시 처음으로 돌아감

            print(f"결과: {num1} {operator} {num2} = {result}")

        except ZeroDivisionError:
            print("에러: 0으로 나눌 수 없습니다.")

        # 5. [조건] 계속 계산할지 확인해야함
        cont = input("계속 계산하시겠습니까? (y/n): ")
        if cont != 'y':
            print("계산기를 종료합니다.")
            break

# 프로그램 실행
run_calculator()
```
