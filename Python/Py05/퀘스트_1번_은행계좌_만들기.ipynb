```python
# [1] random 모듈을 가져옵니다. 계좌번호를 무작위로 생성하기 위해 필요합니다.
import random

# [2] Account라는 이름의 클래스(붕어빵 틀)를 정의합니다.
class Account:
    # [3] 클래스 변수: 클래스 전체에서 공유되는 변수입니다.
    # 생성된 계좌가 총 몇 개인지 저장하는 용도입니다.
    account_count = 0

    # [6] 객체가 생성될 때(Account() 호출 시) 자동으로 실행되는 '생성자' 메서드입니다.
    # self: 나중에 만들어질 '객체 자신'을 가리키는 약속된 기호입니다.
    # name, balance: 사용자가 입력할 '예금주'와 '잔액' 매개변수입니다.
    def __init__(self, name, balance):
        # [7] self.변수명 : 객체 내부에 데이터를 저장합니다.
        self.bank = "SC은행"
        self.name = name
        self.balance = balance

        # [8] random.randint(시작, 끝): 시작~끝 사이의 정수를 랜덤하게 뽑습니다.
        num1 = random.randint(0, 999)
        num2 = random.randint(0, 99)
        num3 = random.randint(0, 999999)

        # [9] f-string: 문자열 안에 변수를 넣을 때 씁니다. {변수:03d}는 '3자리로 만들되 빈 곳은 0으로 채워라'는 뜻입니다.
        # 여기서 '-' 기호는 단순한 구분 문자로 사용되었습니다.
        self.account_number = f"{num1:03d}-{num2:02d}-{num3:06d}"

        # [10] 입금 횟수와 기록을 저장하기 위한 리스트([])와 변수입니다.
        self.deposit_count = 0
        self.deposit_log = [] # []: 빈 리스트를 생성할 때 쓰는 기호입니다.
        self.withdraw_log = []

        # [11] Account.변수명: 클래스 변수에 접근하여 계좌 개수를 1 증가시킵니다.
        Account.account_count += 1

    # [정의] 계좌 개수를 알려주는 클래스 메서드입니다.
    @classmethod
    def get_account_num(cls):
        # cls: 클래스 자체를 의미합니다. (Account와 동일)
        print(f"현재 생성된 계좌 개수: {cls.account_count}개")

    # [14] 입금 기능을 수행하는 메서드입니다.
    def deposit(self, amount):
        # [15] if 조건문: 입금액이 1원 이상인지 확인합니다.
        if amount >= 1:
            self.balance += amount # 기존 잔액에 더합니다.
            self.deposit_log.append(amount) # append(): 리스트의 맨 뒤에 데이터를 추가하는 명령어입니다.
            self.deposit_count += 1

            # [16] %: 나머지를 구하는 기호입니다. 입금 횟수를 5로 나눈 나머지가 0일 때(즉, 5의 배수) 실행됩니다.
            if self.deposit_count % 5 == 0:
                interest = self.balance * 0.01 # 1% 이자 계산
                self.balance += int(interest) # int(): 소수점을 버리고 정수로 만듭니다.
                print(f"** 5회 입금 달성! 이자 {int(interest):,}원이 지급되었습니다. **")
        else:
            print("입금은 최소 1원 이상만 가능합니다.")

    # [정의] 출금 기능을 수행하는 메서드입니다.
    def withdraw(self, amount):
        if self.balance >= amount: # 잔액이 충분할 때만 실행
            self.balance -= amount
            self.withdraw_log.append(amount) # 출금 내역 기록
        else:
            print("잔액이 부족합니다.")

    # [13] 계좌 정보를 화면에 보여주는 메서드입니다.
    def display_info(self):
        print(f"은행이름: {self.bank}")
        print(f"예금주: {self.name}")
        print(f"계좌번호: {self.account_number}")
        # {:,}: 숫자를 세 자리마다 쉼표(,)를 찍어서 출력하라는 서식 지정자입니다.
        print(f"잔고: {self.balance:,}원")

    # [18] 입금 내역을 반복문으로 출력합니다.
    def deposit_history(self):
        print(f"--- {self.name}님의 입금 내역 ---")
        # for 변수 in 리스트: 리스트의 내용을 하나씩 꺼내서 반복합니다.
        for amount in self.deposit_log:
            print(f"입금: {amount:,}원")

# --- 실제 실행 영역 ---

# [4] accounts라는 이름의 빈 리스트를 만듭니다. 여러 계좌를 담을 주머니입니다.
accounts = []

# [5] Account 클래스로 객체를 생성합니다. 이때 [6]번(생성자)으로 점프하여 실행됩니다.
accounts.append(Account("파이썬", 10000))
accounts.append(Account("김철수", 1500000))
accounts.append(Account("이영희", 500000))
accounts.append(Account("박지성", 2000000))

# [12] 리스트 안의 객체들을 하나씩 꺼내어 정보를 확인합니다.
print("=== 잔고 100만원 이상 고객 명단 ===")
for acc in accounts:
    if acc.balance >= 1000000:
        acc.display_info() # [13]번 메서드로 이동
        print("-" * 20)

# [14] 김철수 고객(리스트 1번 인덱스)의 계좌를 가져와 5번 입금해봅니다.
kim = accounts[1]
for _ in range(5): # range(5): 0부터 4까지 5번 반복하라는 뜻입니다.
    kim.deposit(10000) # [15]~[16]번 로직 실행

# [17] 입금 후 최종 정보 출력
kim.display_info()

# [18] 입금 내역 확인
kim.deposit_history()
```

### 붕어빵 틀과 붕어빵 이야기로 배우는 '클래스'와 '객체'

친구가 붕어빵 장사를 한다고 생각해봐요. 붕어빵을 만들려면 뭐가 필요할까요?

1.  **붕어빵 틀 (➡️ 클래스 Class)**: 모양이 똑같은 붕어빵을 계속 만들 수 있는 틀이 필요하겠죠? 이 틀에는 '붕어빵이 어떤 재료로 만들어지는지(속성)'와 '붕어빵을 어떻게 만드는지(기능)'가 정해져 있어요. 파이썬에서는 이 '붕어빵 틀'을 '클래스'라고 불러요.

2.  **붕어빵 (➡️ 객체 Object)**: 붕어빵 틀로 찍어낸 실제 붕어빵 하나하나가 '객체'예요. 모든 붕어빵은 같은 틀에서 나왔지만, 각각 다른 팥이 더 많이 들었을 수도 있고, 좀 더 바삭하게 구워졌을 수도 있죠. 즉, 각 붕어빵(객체)은 자신만의 특별한 특징(데이터)을 가질 수 있어요.

### 우리 노트북의 'Account' 클래스도 붕어빵 틀이에요!

지금 노트북에 `Account`라는 이름의 '클래스'(붕어빵 틀)가 정의되어 있어요. 이 `Account` 클래스는 은행 '계좌'라는 붕어빵을 만드는 틀이에요.

*   **`class Account:`**: 이게 바로 '계좌 붕어빵'을 만들 수 있는 틀을 만든다는 뜻이에요.

*   **`def __init__(self, name, balance):`**: 이 부분은 '붕어빵 틀'에 "어떤 재료가 필요한지"를 알려주는 곳이에요. 새 계좌 붕어빵을 만들 때 '이름'과 '잔액'을 꼭 넣어야 한다는 거죠. 그리고 이 안에서 `self.bank`, `self.name`, `self.balance` 같은 것들이 바로 각 계좌 붕어빵이 가지게 될 자신만의 '속성' (데이터)들이에요.

*   **`def deposit(self, amount):`**, **`def withdraw(self, amount):`**, **`def display_info(self):`**: 이런 부분들은 붕어빵 틀에 "붕어빵이 할 수 있는 일"을 정해놓은 거예요. 즉, 계좌 붕어빵은 '돈을 입금할 수도 있고', '돈을 출금할 수도 있고', '자신의 정보를 보여줄 수도 있다'는 '기능'(메서드)을 가지고 있는 거죠.

### 'Account' 클래스를 이용한 계좌 붕어빵 만들기 예시

아래 코드를 보면 `accounts`라는 리스트에 여러 '계좌 붕어빵'들이 들어있어요. `Account("파이썬", 10000)`처럼 `Account()`를 호출하면 새로운 '계좌 붕어빵'(객체)이 하나씩 만들어지는 거예요.

```python
print('--- 만들어진 계좌 붕어빵들 ---')
for i, acc in enumerate(accounts):
    print(f'{i+1}번째 계좌 붕어빵:')
    acc.display_info() # 각 계좌 붕어빵이 가진 자신만의 정보를 보여줍니다.
    print('-' * 20)

print('\n--- 김철수 붕어빵의 입금 기능 사용 ---')
kim = accounts[1] # 김철수 계좌 붕어빵을 가져옵니다.
kim.deposit(1000) # 김철수 계좌 붕어빵에 돈을 입금하라는 기능을 사용합니다.
kim.display_info() # 입금 후 김철수 계좌 붕어빵의 정보를 다시 보여줍니다.
```
