## Q1-1. 회문 판독기

문자열을 입력으로 받으면 해당 문자열이 회문인지 아닌지 판독하여 회문인 경우 True, 아닌 경우 False를 반환하는 함수를 만듭니다.

이때, 반환되는 True or False는 문자열이 아닌 'bool' 타입입니다.

---
입력 예시:
```python
target1 = "아많다많다많다많아"
target2 = "아정말많다많다많아"

print(is_palindrome(target1))
print(is_palindrome(target2))
```

출력 예시:
```
True
False
```

````python
def is_palindrome(text):
  if text == text[::-1]:
    return True
  else:
    return False
````

````python
# Test
target1 = "여보게저기저게보여"
target2 = "보이긴뭐가보여저기"

print(is_palindrome(target1))
print(is_palindrome(target2))
````

## Q1-2. 회문 분류기

Q1-1에서 만든 판독기 함수를 이용하여 회문 분류기 함수를 만들어 봅니다.

회문 분류기 함수는 아래의 작업을 수행합니다.

1. 여러 단어가 포함된 리스트를 입력으로 받으면 한 단어씩 판독기를 통해 회문 여부를 확인 후 하나의 딕셔너리에 회문과 회문이 아닌 단어를 분류하여 저장합니다.
    - 이때, 회문의 경우 딕셔너리에서 key값은 "회문O", value값은 회문인 단어들로 이루어진 리스트입니다.
    - 회문이 아닌 경우의 key값은 "회문X", value값은 회문이 아닌 단어들로 이루어진 리스트입니다.

2. 모든 단어의 분류가 완료되면 저장한 딕셔너리와, 회문의 총 개수, 전체에서 회문의 비율을 계산하여 반환합니다.
    - 이때, 회문의 총 개수는 int, 회문의 비율은 float입니다.
    - 회문의 비율이 `30%`라면 반환되는 값은 `0.3`입니다.

---
예시 입력:
```python
targets = ['기러기', '까마귀', '토마토', '스윗소스', '스위스', '인도인', '한국인', '별똥별', '우영우', '우향우']

print(palindrome_classifier(targets))
```

예시 출력:
```
({'회문O': ['기러기', '토마토', '스위스', '인도인', '별똥별', '우영우', '우향우'], '회문X': ['까마귀', '스윗소스', '한국인']}, 7, 0.7)
```

````python
def palindrome_classifier(targets):
    회문O = [i for i in targets if is_palindrome(i)]
    회문X = [i for i in targets if not is_palindrome(i)]
    A = len(회문O)
    B = len(회문O)/len(targets)
    return(({'회문O': 회문O, '회문X': 회문X}, A, B))
````

````python
# Test
targets = ['기러기', '까마귀', '토마토', '스윗소스', '스위스', '인도인', '한국인', '별똥별', '우영우', '우향우']

print(palindrome_classifier(targets))
````

## Q2-2. 성적 분석기

이제 Q2-1에서 만든 변수를 참조하여 학생의 이름을 인자로 받아 아래의 예시와 같이 성적을 분석해주는 함수를 만들어 봅니다.

---

예시 입력:
```python
score_analysis("설지민")
```
예시 출력:
```
[설지민 성적 분석]
국어: 88점, 학급 평균(91.50점) 대비 -3.50
영어: 96점, 학급 평균(87.75점) 대비 +8.25
수학: 90점, 학급 평균(87.00점) 대비 +3.00

전과목 평균: 91.33점, 학급 평균(88.75점) 대비 +2.58
```


````python
#해당 코드는 AI를 활용하여 작성하였습니다. (추후 학습용)
def score_analysis(name):
  # score_analysis 함수는 학생의 이름을 인자로 받아 해당 학생의 성적을 분석하고 학급 평균과 비교하여 출력하는 역할을 합니다.

  # 1. 학생 리스트에서 입력받은 이름이 몇 번째(인덱스)에 있는지 찾아요.
  # 'students' 리스트에서 'name' 변수에 해당하는 값의 인덱스(위치)를 찾습니다.
  # 이 인덱스를 사용하여 해당 학생의 점수 데이터에 접근하기 위함입니다.
  # 예: '설지민'은 리스트의 1번에 있습니다. (0부터 시작하니까요!)
  student_index = students.index(name)

  # 2. 찾은 번호를 써서 그 학생의 점수 리스트를 가져와요.
  # 'scores' 리스트는 각 학생의 과목별 점수를 담고 있으며, 위에서 찾은 'student_index'를 이용해
  # 현재 분석하려는 학생의 국어, 영어, 수학 점수 리스트를 'my_scores' 변수에 할당합니다.
  # my_scores는 [88, 96, 90]이 됩니다.
  my_scores = scores[student_index]

  # 3. 학급 전체의 과목별 평균을 구해요.
  # 'students' 리스트의 길이를 통해 전체 학생 수를 파악합니다.
  num_students = len(students)

  # 국어(0번), 영어(1번), 수학(2번) 점수들을 다 합쳐서 인원수로 나눠요.
  # 각 과목별로 모든 학생의 점수를 더한 후 총 학생 수로 나누어 학급 평균을 계산합니다.
  # s[0]은 첫 번째 과목(국어), s[1]은 두 번째 과목(영어), s[2]는 세 번째 과목(수학)의 점수를 의미합니다.
  avg_kor = sum(s[0] for s in scores) / num_students # 국어 학급 평균 계산
  avg_eng = sum(s[1] for s in scores) / num_students # 영어 학급 평균 계산
  avg_math = sum(s[2] for s in scores) / num_students # 수학 학급 평균 계산

  # 학급 전체 평균 (세 과목 평균의 평균)
  # 계산된 각 과목의 학급 평균을 모두 더한 후 과목 수(3)로 나누어 학급 전체의 평균 점수를 계산합니다.
  avg_class_total = (avg_kor + avg_eng + avg_math) / 3

  # 4. 학생 개인의 평균도 구해요.
  # 'my_scores'에 담긴 현재 학생의 모든 과목 점수를 더한 후 과목 수(3)로 나누어 개인 평균을 계산합니다.
  my_avg = sum(my_scores) / 3

  # 결과 출력 부분:
  # print() 함수 앞에 'f'가 붙은 것은 F-string(Formatted string literals)입니다.
  # F-string을 사용하면 문자열 안에 변수나 표현식의 값을 {}(중괄호) 안에 넣어 바로 출력할 수 있어 가독성이 높습니다.
  # 예: f"[{name} 성적 분석]"은 'name' 변수의 값이 문자열 안에 삽입됩니다.
  print(f"[{name} 성적 분석]") # 학생 이름과 함께 성적 분석 시작을 알림
  print(f"국어: {my_scores[0]}점, 학급 평균({avg_kor:.2f}점) 대비 {my_scores[0] - avg_kor:+.2f}")
  # my_scores[0]: 해당 학생의 국어 점수
  # avg_kor:.2f: 학급 국어 평균을 소수점 둘째 자리까지 표시
  # my_scores[0] - avg_kor:+.2f: 학생 점수와 학급 평균의 차이를 계산하고, 양수일 경우 '+' 부호를 붙여 소수점 둘째 자리까지 표시
  print(f"영어: {my_scores[1]}점, 학급 평균({avg_eng:.2f}점) 대비 {my_scores[1] - avg_eng:+.2f}")
  # 위와 동일하게 해당 학생의 영어 점수와 학급 평균 및 차이를 출력
  print(f"수학: {my_scores[2]}점, 학급 평균({avg_math:.2f}점) 대비 {my_scores[2] - avg_math:+.2f}")
  # 위와 동일하게 해당 학생의 수학 점수와 학급 평균 및 차이를 출력
  print("") # 가독성을 위해 한 줄 띄우기
  print(f"전과목 평균: {my_avg:.2f}점, 학급 평균({avg_class_total:.2f}점) 대비 {my_avg - avg_class_total:+.2f}")
  # 학생의 전과목 평균과 학급 전체 평균 및 그 차이를 출력
````

````python
score_analysis("설지민")



````
